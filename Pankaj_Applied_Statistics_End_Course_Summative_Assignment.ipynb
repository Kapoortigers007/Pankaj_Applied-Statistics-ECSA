### **60. What is the difference between Descriptive and Inferential Statistics?**
Descriptive Statistics:
Concerned with summarizing and organizing data. It describes the main features of a dataset using measures like mean, median, mode, variance, standard deviation, tables, and graphs.

Inferential Statistics:
Concerned with drawing conclusions or making predictions about a population based on a sample. It uses methods such as confidence intervals, hypothesis testing, regression, and correlation.

In short:
Descriptive statistics describe what the data shows, while inferential statistics infer conclusions beyond the observed data
